```
import ee
```

```
ee.Authenticate()
```

```
ee.Initialize()
```

```
geo = ee.Geometry.Polygon([
    [
      [
        -97.2829761095046,
        46.281829508788256
      ],
      [
        -98.01002984732004,
        46.28131183102219
      ],
      [
        -98.04288956396547,
        46.282292222468485
      ],
      [
        -98.0366477470581,
        46.629178856849336
      ],
      [
        -97.68405608922647,
        46.63081715959645
      ],
      [
        -97.28371726989675,
        46.63150869171885
      ],
      [
        -97.2829761095046,
        46.281829508788256
      ]
    ]
  ]);
```

```
collection = ee.ImageCollection('LANDSAT/LC08/C01/T1_RT'
).filterDate('2018-07-01', '2018-09-15'
).filterBounds(geo).select(['B4', 'B3', 'B2']).filter(ee.Filter.lt('CLOUD_COVER', 5));
```

```
collection_list = collection.toList(collection.size())
collection_size = collection_list.size().getInfo()
for i in range(collection_size):
    ee.batch.Export.image.toDrive(
        image = ee.Image(collection_list.get(i)), #.clip(geo), 
        fileNamePrefix = 'landsat_' + str(i + 1) , 
        folder = 'landsat_8'
        #dimensions = '128x128'
        ).start()
    print('exporting image {}'.format(i))
```
